```python
import pandas as pd

# Cargar el dataset

df = pd.read_csv("test.csv")

# Consulto el tamaño del dataset


print("Tamaño:", df.shape)
print("")

# Cuento los valores nulos del dataset haciendo una suma de ellos por columnas, para ver en cada parametro cuantos datos no cuentan

print("Datos nulos por columna:")
print(df.isnull().sum())
print("")

# Veo los tipos de datos de cada columna

print("Tipos de datos:")
print(df.dtypes)
print("")

# Estudiamos el tipo de datos de cada columna para saber si son cualitativas o cuantitativas
# En este caso, como son 11 columnas no haria falta utilizar programación para analizarlo sin embargo este sería el procedimiento

cualitativas = df.select_dtypes(include=['object']).columns.tolist()
cuantitativas = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
print("Las variables cualitativas generalmente son aquellas cuyo tipo de dato es 'object':", cualitativas)
print("Las variables cuantitativas generalmente son aquellas cuyo tipo de dato es 'int64' o 'float64':", cuantitativas)
print("")

# Podemos ver una columna aparece como ..., eso indica que pandas está truncando la visualización de los tipos de datos 
# debido a que el DataFrame tiene muchas columnas o el nombre de las columnas es demasiado largo
# Estas son algunas soluciones ante este problema

# Ver el tipo de datos de una columna específica
print("El tipo de datos de la columna Cabin es: ", df['Cabin'].dtype)
print("")


# Convertir a DataFrame y mostrar sin truncar
dtypes_df = df.dtypes.reset_index()
dtypes_df.columns = ['Column Name', 'Data Type']
print(dtypes_df.to_string(index=False))
print("")

# Cambiar la configuración de pandas para mostrar más columnas
pd.set_option('display.max_columns', None)  # Muestra todas las columnas
```
